```python
def get_range(x):
    if x['Effective_Price'] > 500:
        return 1
    else:
        return 0.5
```

```python
data['Range'] = data.apply(get_range, axis=1)
```

```python
print(data)
```

```python
imputer = SimpleImputer(strategy='mean')
data['Percent_Quantity_With_Discount'] = imputer.fit_transform(data[['Percent_Quantity_With_Discount']])
```

```python
encoder = OneHotEncoder(sparse=False)
mapped_sales_type_encoded = encoder.fit_transform(data[['Mapped_Sales_Type']])
mapped_sales_type_encoded_data = pd.DataFrame(mapped_sales_type_encoded, columns=encoder.get_feature_names(['Mapped_Sales_Type']))
data = pd.concat([data, mapped_sales_type_encoded_data], axis=1)
```

```python
binarizer = LabelBinarizer()
day_binarized = binarizer.fit_transform(data['Day'])
data['Day_binarized'] = day_binarized
```

```python
label_encoder = LabelEncoder()
data['City_encoded'] = label_encoder.fit_transform(data['City'])
```
